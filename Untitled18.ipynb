```python
import math
def modulus(a,b,p):
    inv=0
    num=1
    while((b*num)%p!=1):
        num+=1
    inv=num
    return ((a%p)*inv)%p


E=list(map(int,input().split(' ')))
G=list(map(int,input().split(' ')))
p=11
a,b=E[0],E[1]

xp,yp=G[0],G[1]

xq,yq=xp,yp


lbd=modulus((3*xp*xp+a),(2*yp),p)

print("Lambda = ",lbd)


xr=(lbd*lbd-xp-xq)%p
yr=(lbd*(xp-xr)-yp)%p

print("xr = ",xr," yr = ",yr)

Gn=[xr,yr]

print("2 G"," = ",Gn[0]," ",Gn[1])

count=3

while(count!=14):
    xp,yp=Gn[0],Gn[1]
    xq,yq=G[0],G[1]
    lbd=modulus(yq-yp,xq-xp,p)
    print("\nlambda = ","(",yq,"-",yp,")","/","(",xq,"-",yq,") mod",p)
    print("lambda = ",lbd)
    xr=(lbd*lbd-xp-xq)%p
    print("xr = (",lbd*lbd,"-",xp,"-",xq,") mod ",p)
    print("xr = ",xr)
    yr=(lbd*(xp-xr)-yp)%p
    print("yr = (",lbd,"(",xp,"-",xr,")-",yp,")mod",p)
    print("yr = ",yr)
    Gn=[xr,yr]
    print(count,"G"," = ",Gn[0]," ",Gn[1])
    count+=1
```
