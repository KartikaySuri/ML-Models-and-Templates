```python
import tensorflow as tf
import pandas as pd
import numpy as np
import tensorflow_hub as hub
import matplotlib.pyplot as plt
```

```python
train_dir = pd.read_csv("/content/train.csv")

test_dir = pd.read_csv("/content/test.csv")
```

```python
train_df_shuffle = train_dir.sample(frac = 1, random_state = 42)
train_df_shuffle.head()
```

```python
train_df_shuffle.target.value_counts()
```

```python
from sklearn.model_selection import train_test_split
train_sentence, val_sentence, train_label, val_label = train_test_split(train_df_shuffle["text"],
                                                                        train_df_shuffle["target"],
                                                                        random_state = 42,
                                                                        test_size =0.2)
```

```python
import tensorflow_hub as hub
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
embed_sample = embed(["this is a sample sentence"])

print(embed_sample)
```

```python
sentence_encoder_layer = hub.KerasLayer("https://tfhub.dev/google/universal-sentence-encoder/4",
                                        input_shape = [],
                                        dtype = tf.string,
                                        trainable = False,
                                        name = "USE")
```

```python
from tensorflow.keras import layers

model_0 = tf.keras.Sequential([
    sentence_encoder_layer,
    layers.Dense(64, activation = "relu"),
    layers.Dense(1, activation = "sigmoid")
], name = "model_0")
```

```python
model_0.compile(loss = "binary_crossentropy",
                optimizer = "Adam",
                metrics = ["accuracy"])
```

```python
history_0 = model_0.fit(train_sentence,
                        train_label,
                        epochs = 5,
                        validation_data = (val_sentence, val_label))
```

```python
model_0.save("test.h5")
```
