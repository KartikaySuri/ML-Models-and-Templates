```python
#Kartikay Suri 20BCE0849
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset = pd.read_csv('us-marriages-divorces-1867-2014.csv')
x = dataset.iloc[:,0].values #complete set of years
y1 = dataset.iloc[:,-2].values # complete set of marriages
y2 = dataset.iloc[:,-1].values # complete set of deaths
plt.plot(x,y1,label = 'Marriages per 100' )
plt.plot(x,y2,label = 'Divorce per 100')
plt.legend(loc='best')
plt.xlabel("Year")
plt.ylabel("Per capita value")
plt.title("Marriage and Death per capita in us")
plt.figure()
plt.show()
```

```python
#Kartikay Suri (20BCE0849)
dataset = pd.read_csv("us-marriages-divorces-1867-2014.csv")
x = dataset.iloc[124:134,0].values #complete set of years from 1991 to 2000
y1 = dataset.iloc[124:134,-2].values # complete set of marriages
y2 = dataset.iloc[124:134,-1].values # complete set of deaths
X_axis = np.arange(len(x))
plt.bar(X_axis-0.2, y1, 0.4, label = 'Marriages per 100')
plt.bar(X_axis+0.2, y2, 0.4, label = 'Divorces per 100')
  
plt.xticks(X_axis, x)
plt.xlabel("Years")
plt.ylabel("Number per 100")
plt.title("Per capita marriage and divorce comparison between 1991 and 2000")
plt.legend()
plt.show()
```
