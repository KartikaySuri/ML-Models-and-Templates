```python
# performing matrox multiplication on tneosrs in tf
tfmul1 = tf.constant([[10,7],
                      [7,10]])
tfmul2 = tf.constant([[1,1],
                      [1,1]])
tf.matmul(tfmul1,tfmul2)
```

```python
tfmul3 = tf.constant([[[1,2,5],
                       [7,2,1],
                       [3,3,3]]])

tfmul4 = tf.constant([[3,5],
                      [6,7],
                      [1,8]])

tf.matmul(tfmul3,tfmul4)
```

```python
tfmul3 @ tfmul3
```
