```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import imutils
import easyocr
import time
```

```python
cap = cv2.VideoCapture(0)
pTime,cTime = 0,0
while True:
  success,img = cv2.imread('image4.jpg')
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
  cTime = time.time()
  fps = 1/(cTime-pTime)
  pTime = cTime
  cv2.putText(cv2.flip(img,1),str(int(fps)),(10,70),cv2.FONT_HERSHEY_PLAIN,3,(225,0,225),3)
  cv2.imshow("Image", cv2.flip(img,1))
  cv2.waitKey(1)
```
