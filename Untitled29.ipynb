```python
import pandas as pd
import numpy as np
import pickle
import joblib
from sklearn.feature_extraction import DictVectorizer
```

```python
df = pd.read_csv('height_and_pose.csv')
featval,target,weights = [],[],[]

for i in range(len(df)):
  dfrow,dftarget = df.iloc[i,-1],df.iloc[i,1]
  dfrow = eval(dfrow[1:-1])
  emptdict = {}
  emptweight = []
  for j in dfrow['key_points_coordinate']:
    keyset,valset = list(j.keys())[0],list(j.values())[0]
    emptdict[(str(keyset) + '_x')] = valset['x']
    emptdict[(str(keyset) + '_y')] = valset['y']
  for j in dfrow['key_points_prob']:
    emptweight.append(list(j.values())[0]['score'])
    emptweight.append(list(j.values())[0]['score'])
  featval.append(emptdict)
  target.append(dftarget)
  sample_weights = np.mean(emptweight, axis=0) 
  weights.append(sample_weights)
```

```python
for i in featval:
  print(i)
```

```python
vec = DictVectorizer()
print("Feature Matrix: ")
print(vec.fit_transform(featval).toarray())
featvalarr = vec.fit_transform(featval).toarray()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(featvalarr, target, test_size=0.2, random_state=42)
```

```python
model1 = DecisionTreeRegressor(max_depth=10, min_samples_split=2, min_samples_leaf=2)
model2 = DecisionTreeRegressor(max_depth=5)

model1.fit(X_train, y_train,sample_weights)
model2.fit(X_train, y_train)

y_pred1 = model1.predict(X_test)
y_pred2 = model2.predict(X_test)
print(y_pred1,y_pred2)
print(y_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.utils import load_img, img_to_array
from keras.applications.vgg16 import preprocess_input, VGG16
import os
 
path = "/content/depthmap/"
image_paths = []
targetvals = []
for i in range(len(df)-1):
  targetvals.append(df.iloc[i,1])
  image_paths.append(path +str(df.iloc[i,0])+'.jpg')
print(image_paths)
print(targetvals)
targetvals = np.array(targetvals) 
```

```python
images = []
for path in image_paths:
    img = load_img(path, target_size=(224, 224)) 
    img_arr = img_to_array(img) 
    img_arr = preprocess_input(img_arr) 
    images.append(img_arr)

X = np.array(images)
model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

features = model.predict(X)
X_flat = features.reshape(features.shape[0], -1)

```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
regressor.fit(X_flat, targetvals)
```
