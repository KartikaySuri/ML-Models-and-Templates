```python
# from ctypes import LibraryLoader
# # typical architecture for  convolutional neural network
# input image along with the input shape
# input layer for the nn with the same input shape
# followed by convolutional layer which also has a hidden activation function
# after which we will proceed to add a pooling layer to reduce the dimensionality of the network
# then we will add a fully connected layer with another hidden activation function
# we will finally end with the fully connected layers connected to a dense layer containing the output and activation based on the classification
# problem and the number of classes to output
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

tf.random.set_seed(42)
train_datagen = ImageDataGenerator(rescale = 1/255.)
test_datagen = ImageDataGenerator(rescale = 1/255.)

train_dir = "/content/pizza_steak/train"
test_dir = "/content/pizza_steak/test"

train_data = train_datagen.flow_from_directory(train_dir,
                                               batch_size = 32,
                                               target_size = (224,224),
                                               class_mode = "binary",
                                               seed = 42)
valid_data = test_datagen.flow_from_directory(test_dir,
                                             batch_size = 32,
                                             class_mode = "binary",
                                             target_size = (224,224),
                                             seed = 42)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, Dense, GlobalAvgPool2D, MaxPool2D, Flatten
```

```python
model_0 = Sequential([
    Conv2D(filters = 10,
           kernel_size = 3,
           activation = "relu",
           input_shape = (224,224,3)),
    Conv2D(10,3, activation = "relu"),
    MaxPool2D(pool_size = 2,
              padding = "valid"),
    Conv2D(10,3, activation = "relu"),
    Conv2D(10,3, activation = "relu"),
    MaxPool2D(2),
    Flatten(),
    Dense(1, activation = "sigmoid")
])
```

```python
model_0.compile(loss = tf.keras.losses.BinaryCrossentropy(),
                optimizer = tf.keras.optimizers.Adam(),
                metrics = ["accuracy"])
```

```python
history_0 = model_0.fit(train_data,
                        epochs = 5,
                        steps_per_epoch = len(train_data),
                        validation_data = valid_data,
                        validation_steps = int(0.25 * len(valid_data)))
```

```python
import tensorflow as tf
def create_tensorboard_callback(dir_name, experiment_name):
  log_dir = dir_name + "/" + experiment_name + "/" + datetime.datetime.now().strftime("%Y%m%d - %H%M%S")
  tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir = log_dir)
  return tensorboard_callback
```

```python
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras import layers
```

```python
resnet_url = "https://tfhub.dev/google/imagenet/resnet_v2_50/feature_vector/4"

# Original: EfficientNetB0 feature vector (version 1)
efficientnet_url = "https://tfhub.dev/tensorflow/efficientnet/b0/feature-vector/1"
```

```python
def create_model(model_url, num_classes = 10):
  feature_extractor_layer = hub.KerasLayer(model_url,
                                           trainable = False,
                                           name = "feature_extraction",
                                           input_shape = (224,224,3))

  model = tf.keras.Sequential([
      feature_extractor_layer,
      layers.Dense(num_classes, activation = "softmax")
  ])

  return model
```

```python
resnet_model = create_model(resnet_url)
effecient_net = create_model(efficientnet_url)
```

```python
from tensorflow.python import train
resnet_model.compile(loss = "categorical_crossentropy",
                     optimizer = "Adam",
                     metrics = ["accuracy"])

resnet_history = resnet_model.fit(train_data,
                                  epochs = 5,
                                  steps_per_epoch = len(train_data),
                                  validation_data = test_data,
                                  validation_steps = int(0.25 * len(test_data)),
                                  callbacks = [create_tensorboard_callback("trained_model","resnet")])
```

```python
effecient_net.compile(loss = "categorical_crossentropy",
                     optimizer = "Adam",
                     metrics = ["accuracy"])

effecient_net_history = effecient_net.fit(train_data,
                                          epochs = 5,
                                          steps_per_epoch = len(train_data),
                                          validation_data = test_data,
                                          validation_steps = int(0.25 * len(test_data)),
                                          callbacks = [create_tensorboard_callback("trained_model","effecientnet")])
```
