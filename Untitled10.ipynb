```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df1 = pd.read_csv("recent-college-grads-earnings.csv")
df2 = pd.read_csv("us-marriages-divorces-1867-2014.csv")
df3 = pd.read_csv("percent-degrees-conferred-women-usa.csv")
```

```python
men_p25,men_med,men_p75,men_field = [],[],[],[]
women_p25,women_med,women_p75,women_field= [],[],[],[]
for i in range(len(df1)):
  if df1.loc[i,"Men"]>df1.loc[i,"Women"]:
    men_p25.append(df1.loc[i,"P25th"])
    men_med.append(df1.loc[i,"Median"])
    men_p75.append(df1.loc[i,"P75th"])
    men_field.append(df1.loc[i,"Major"])
  else:
    women_p25.append(df1.loc[i,"P25th"])
    women_med.append(df1.loc[i,"Median"])
    women_p75.append(df1.loc[i,"P75th"])
    women_field.append(df1.loc[i,"Major"])
```

```python
data_men = [men_p25,men_med,men_p75]
data_women = [women_p25,women_med,women_p75] 
fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(data_men)
ax.set_xticklabels(['P25th', 'median','P75th'])
fig = plt.figure(figsize = (10,7))
ax2 = fig.add_axes([0,0,1,1])
bp2 = ax2.boxplot(data_women)
ax2.set_xticklabels(['P25th', 'median','P75th'])
plt.show()
```

```python
marr_percap,div_percap = [],[]
year = df2.Year
for i in range(len(df2)):
  percapdiv = (df2.loc[i,"Marriages"]/df2.loc[i,"Population"])
  percapmar = (df2.loc[i,"Divorces"]/df2.loc[i,"Population"])
  marr_percap.append(percapmar)
  div_percap.append(percapdiv)
dist1 = marr_percap
dist2 = div_percap
fig, axs = plt.subplots(2)
axs[0].plot(year, dist1)
axs[1].plot(year, dist2)
plt.xlabel('Year')
```

```python
plt.figure(figsize=(20, 10))
fields = list(df3.columns)
year = df3.Year
for i in range(0,17):
  plt.subplot(3, 6, i+1)
  plt.title(i)
  value = fields[i+1]
  y_vals = df3[value]
  plt.plot(year, y_vals)
  plt.ylim(0,90)
  plt.title('University {}'.format(i+1))
  
```
