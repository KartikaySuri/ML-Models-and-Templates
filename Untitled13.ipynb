```python
#Grouped Bar Chart

import plotly.express as px
df = px.data.iris()
fig = px.bar(df, x="petal_width", y="petal_length",color="species", 
             hover_data=['sepal_width','sepal_length'],
             barmode = 'group')
fig.show()
```

```python
#Scatter Plot
import plotly.express as px
dataframe = px.data.iris()
fig = px.scatter(dataframe, x="species", y="sepal_width", color="species")
fig.show()
```

```python
#Histogram
import plotly.express as px
  
df = px.data.iris() 
fig = px.histogram(df, x=["petal_length"])
fig.show()
```

```python
#Pie Chart
import plotly.express as px
 
dataframe = px.data.iris()
fig = px.pie(dataframe, values="petal_width", names="species",
             title='Iris Dataset', hover_data=['sepal_length',
                                               'sepal_width',
                                               'petal_length'])
fig.show()
```

```python
#Box Plot
import plotly.express as px
 
dataframe = px.data.iris()
fig = px.box(dataframe, x="petal_width", y="petal_length",
             hover_data=['sepal_length','sepal_width'],
             color = 'species')
fig.show()
```

```python
#Donut Plot
import plotly.graph_objects as go

dataframe = px.data.iris()
fig = go.Figure(data=[go.Pie(labels=dataframe['species'], 
                             values=dataframe['petal_length'],
                             hole=.4)])
fig.show()
```
