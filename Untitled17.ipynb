```python
# client.py
import time, socket, sys
print("\nWelcome to Chat Room\n")
print("Initialising....\n")
time.sleep(1)
s = socket.socket(socket.AF_INET,socket.SOCK_DGRAM)
shost = socket.gethostname()
ip = socket.gethostbyname(shost)
print(shost, "(", ip, ")\n")
host = input(str("Enter server address: "))
name = input(str("\nEnter your name: "))
port = 1234
print("\nTrying to connect to ", host, "(", port, ")\n")
time.sleep(1)
s.connect((host, port))
print("Connected...\n")
s.sendto(name.encode(),(host,port))
print("bruh")
s_name = s.recvfrom(1024)
print(s_name)
s_name = s_name.decode()
print("bruh")
print(s_name, "has joined the chat room\nEnter [e] to exit chat room\n")
while True:
    message = s.recvfrom(1024)
    #message = message.decode()
    print(s_name, ":", message)
    message = input(str("Me : "))
    if message == "[e]":
        message = "Left chat room!"
        s.sendto(message.encode(),(host,port))
        print("\n")
        break
    s.sendto(message.encode(),(host,port))

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.layers import Dense, Flatten
from keras.models import Model
from keras.applications.inception_v3 import InceptionV3, preprocess_input
from keras.preprocessing.image import ImageDataGenerator
import keras
```

```python
base_model = InceptionV3(input_shape =(225,225,3), include_top = False )
```

```python
for layer in base_model.layers:
  layer.trainable = False
  
```

```python
x = Flatten()(base_model.output)
x = Dense(units = 11, activation = 'sigmoid')(x)

model = Model(base_model.input,x)
model.compile(optimizer = 'adam', loss = keras.losses.binary_crossentropy,metrics = ['accuracy'])
```

```python
train_datagen = ImageDataGenerator(featurewise_center= True,
rotation_range=0.4,
width_shift_range=0.3,
horizontal_flip=True,
preprocessing_function=preprocess_input,
zoom_range=0.4,shear_range=0.4)

train_data = train_datagen.flow_from_directory(directory = '/content/drive/MyDrive/dataset',
target_size=(225,225),
batch_size = 32)
```

```python
def plotImages(img_arr, label):
    for idx, img in enumerate(img_arr):
        if idx<=10:
            plt.figure(figsize = (5,5))
            plt.imshow(img)
            plt.title(img.shape)
            plt.axis = False
            plt.show()
```

```python
plotImages(t_img,label)
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping
mc = ModelCheckpoint(filepath='./best_model.h5', 
monitor = 'accuracy',
verbose = 1,
save_best_only=True)

# Es = EarlyStopping(monitor="accuracy",
# min_delta = 0.01, 
# patience=5, 
# verbose=1)

cb = [mc]
```

```python
his = model.fit_generator(train_data, 
steps_per_epoch = 10, 
epochs = 30, callbacks = cb)
```
