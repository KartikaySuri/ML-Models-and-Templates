# finding the best learning rate
to find the learning rate where the loss ecreases the most during training, we will be using the following steps:

firstly we will use a learning rate callback
call back is an extra piece of functionality which we can add to the model while it is training

another model, we could use the same one as above but since we are practicing building the model we will be using a fresh model

finally, using a modified loss curve


```python
tf.random.set_seed(42)


model6 = tf.keras.Sequential([
    tf.keras.layers.Dense(4,activation = "relu"),
    tf.keras.layers.Dense(4,activation = "relu"),
    tf.keras.layers.Dense(1,activation = "sigmoid")
])

model6.compile(loss = "binary_crossentropy",
               optimizer = "Adam",
               metrics = ["Accuracy"])

# creating a learning rate callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10 ** (epoch/20))
history = model6.fit(Xtrain,Ytrain,
                     epochs = 101,
                     callbacks = [lr_scheduler])
```

```python
pd.DataFrame(history.history).plot(figsize = (12,6),xlabel = "epochs")
```

```python
# plot the learning rates versus the loss we have observed in the training
lrs = 1e-4 * (10 **(tf.range(101)/20))
plt.figure(figsize = (10,7))
plt.semilogx(lrs,history.history["loss"])

plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning Rate vs Loss")
```

```python
tf.random.set_seed(42)

model9 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape = (28,28)),
    tf.keras.layers.Dense(4,activation = "relu"),
    tf.keras.layers.Dense(4,activation = "relu"),
    tf.keras.layers.Dense(10, activation = "softmax")
])


model9.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy(),
               optimizer = tf.keras.optimizers.Adam(),
               metrics = ["Accuracy"])

lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-3 * 10 **(epoch/20))

history = model9.fit(train_data_norm,train_label,epochs = 40, validation_data = (test_data_norm,test_label), callbacks = [lr_scheduler])
```

```python
import numpy as np
import matplotlib.pyplot as plt

lrs = 1e-3 * 10 ** (tf.range(40)/20)
plt.semilogx(lrs,history.history["loss"])
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Finding the idle learning rate")
```

```python
# refitting the model with the idle learning rate
tf.random.set_seed(42)

model10 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape = (28,28)),
    tf.keras.layers.Dense(4,activation = "relu"),
    tf.keras.layers.Dense(4,activation = "relu"),
    tf.keras.layers.Dense(10,activation = "softmax")
])

model10.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy(),
                optimizer = tf.keras.optimizers.Adam(lr = 0.001),
                metrics = ["Accuracy"])

model10.fit(train_data_norm,train_label,epochs= 40, validation_data = (test_data_norm,test_label))
```

```python
# evaluating the classification model

#  plotting the confusion matrix using the default function
import itertools
from sklearn.metrics import confusion_matrix
def plotmatrix(Xtest,Ytest,model7, classes = None,figsize = (10,10)):

  figsize = (10,10)

  cm = confusion_matrix(Ytest,tf.round(model7.predict(Xtest)))

  cm_norm = cm.astype("float")/cm.sum(axis = 1)[:,np.newaxis]

  n_classes = cm.shape[0]

  fig,ax = plt.subplots(figsize = figsize)

  cax = ax.matshow(cm,cmap = plt.cm.Blues)

  fig.colorbar(cax)

  classes = False

  if classes:
    labels = classes
  else:
    labels = np.arange(cm.shape[0])

  ax.set(title = "confusion matrix",
          xlabel = "predicted",
          ylabel = "true",
          xticks = np.arange(n_classes),
          yticks = np.arange(n_classes),
          xticklabels = labels,
          yticklabels = labels
          )
  threshold = (cm.max() + cm.min())/2
  for i,j in itertools.product(range(cm.shape[0]),range(cm.shape[1])):
    plt.text(j,i,f"{cm[i,j]}({cm_norm[i,j]*100:.1f}%)",
            horizontalalignment = "center",
            color="white" if cm[i,j] > threshold else "black",
            size = 15
            )
```
